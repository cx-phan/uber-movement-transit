```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import folium
from folium import plugins
import data
```

```python
TIME_PATH = 'san_francisco-censustracts-2018-1-All-HourlyAggregate.csv'
COORD_PATH = 'san_francisco_censustracts.json'
COORD_ID = 2308
HOD = 12
```

```python
df = pd.read_csv(TIME_PATH)
coord_dict = data.get_coord(COORD_PATH)
```

```python
# PRINT TOP NODES
# df['sourceid'].value_counts()
```

```python
# GET HEATMAP DATA

rows = df.loc[df['sourceid'] == COORD_ID]
max_time = max(rows['mean_travel_time'])

heat_data = []
for r in rows.iterrows():
    dstid = r[1]['dstid']
    coord = coord_dict[dstid][::-1]
    time = r[1]['mean_travel_time']
    
    if r[1]['hod'] != HOD: continue
    heat_data.append([coord[0], coord[1], time])

heat_data = np.array(heat_data)
print(heat_data.shape)
heat_data[:, 2] /= np.max(heat_data[:, 2])
```

```python
# ADD MARKER TO LOCATION
location = coord_dict[COORD_ID][::-1]
map_sf = folium.Map(location=location,
                    zoom_start=10)
marker = folium.Marker(location)
marker.add_to(map_sf)
```

```python
hm = plugins.HeatMap(heat_data, min_opacity=0, radius=20, max_zoom=10)
hm.add_to(map_sf)

fullscreen = folium.plugins.Fullscreen()
fullscreen.add_to(map_sf)
```
